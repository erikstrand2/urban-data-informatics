# Scraping text from the webpage
### 1. Let's get text information when searching 'urban planning' in the New York Times website.

```python
url = 'https://www.nytimes.com/search?query=urban+planning'
response = urllib.request.urlopen(url)
html = response.read()
mystr = html.decode("utf8")
response.close()
print (mystr)
```

### 2. Text mining using BeautifulSoup libarary

```python
soup = BeautifulSoup(mystr,"html5lib")
```

### 3. Retrieve relevant information. We would like to find the following information for each arcitle.
* Headline
* Summary
* Article category
* Author
### Use _find_all( )_ function to find information with the same tag.

```python
# Text with a tag 'h4', which is a headline
soup.find_all('h4')
```

```python
# Text with a tag 'p', including summaries, categories, and authors
soup.find_all('p')
```

```python
# Let's just find text about article summaries.
soup.find_all('p', attrs={'class':'css-16nhkrn'})
```

### 4. Extract relevant information and store in Python lists

```python
result_headlines = soup.find_all('h4')
result_summaries = soup.find_all('p', attrs={'class':'css-16nhkrn'})
result_categories = soup.find_all('p', attrs={'class':'css-myxawk'})
result_authors = soup.find_all('p', attrs={'class':'css-15w69y9'})

print (result_headlines[0])
print (result_summaries[0])
print (result_categories[0])
print (result_authors[0])
```

```python
# Remove any unnecessary information (e.g. <h4 class="css-2fgx4k">)
result_headlines = [i.text.strip() for i in result_headlines]
result_summaries = [i.text.strip() for i in result_summaries]
result_categories = [i.text.strip() for i in result_categories]
result_authors = [i.text.strip() for i in result_authors]
```

```python
result_headlines
```

### 5. Create dataframe

```python
NYT_result = pd.DataFrame()
NYT_result['headline'] = result_headlines
NYT_result['summary'] = result_summaries
NYT_result['category'] = result_categories
NYT_result['author'] = result_authors
NYT_result.head()
```

# TripAdvisor Example:
## How to get additional information from the next page using BeautifulSoup?

```python
# Create an empty list
URLs = []

# Put the first item (the first page of the search results) manually
URLs.append('https://www.tripadvisor.com/Attractions-g60763-Activities-c61-New_York_City_New_York.html')

# Using for loops to append valid URLs automatically (recursive process)
# %i is replaced every cycle (30, 60, 90, ... 330)
for i in range(30,360,30): 
    URLs.append('https://www.tripadvisor.com/Attractions-g60763-Activities-c61-oa%i-New_York_City_New_York.html'%(i))
    
URLs
```

```python
# Create empty lists to store information 
# Every cycle of the for loop will append information to this list.

outdoor_activities_total = []
reviews_total = []

for url in URLs:
    response = urllib.request.urlopen(url)
    html = response.read()
    mystr = html.decode("utf8")
    response.close()

    soup = BeautifulSoup(mystr,"html5lib")

    # Outdoor activities
    outdoor_activities = soup.find_all('h2') # BeautifulSoup type data
    outdoor_activities = [i.text.strip() for i in outdoor_activities] # convert to text object
    
    # .extend() function is for appending and flattening multiple lists
    outdoor_activities_total.extend(outdoor_activities)
    
outdoor_activities_total[:5] # first five (5) items
```

```python
# Then you can create Pandas dataframe to create a tabular data
NYC_outdoor = pd.DataFrame() # create an empty dataframe
NYC_outdoor['activity'] = outdoor_activities_total

print (len(NYC_outdoor))
NYC_outdoor.head()
```
